```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Nomor 1

```python
df_1 = df.describe()

df_1.drop(columns = 'Overall rank', inplace=True)

df_1.loc["IQR"] = df_1.loc['75%'] - df_1.loc['25%']

df_1.loc["Range"] = df_1.loc['max'] - df_1.loc['min']

df_1.loc["Variance"] = df_1.loc["std"] ** 2

df_1
```

```python
df_zscore = df.copy()

df_zscore.drop(columns = ['Overall rank', 'Country or region'], inplace=True)

cols = list(df_zscore.columns)

for col in cols:
    col_zscore = col + '_zscore'
    df_zscore[col_zscore] = (df_zscore[col] - df_zscore[col].mean())/df_zscore[col].std(ddof=1)
df_zscore
```

## Nomor 2

```python
df_2 = df_1.copy()

df_2.loc['High_Out_IQR'] = df_2.loc['75%'] + (df_2.loc["IQR"] * 1.5)

df_2.loc['Low_Out_IQR'] = df_2.loc['25%'] - (df_2.loc["IQR"] * 1.5)

df_2.loc["High_Out_Zscore"] = df_2.loc["mean"] + (df_2.loc["std"] * 2.5)

df_2.loc["Low_Out_Zscore"] = df_2.loc["mean"] - (df_2.loc["std"] * 2.5)

df_2
```
